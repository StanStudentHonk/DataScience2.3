```python
import pandas as pd
```

```python
import seaborn as sns
```

```python
lifeExpectancy = pd.read_csv('lifeExpectancyAtBirth.csv', sep=',')
```

```python
lifeExpectancy
```

```python
lifeExpectancy2019AndBothSexes = (lifeExpectancy['Period'] == 2019) & (lifeExpectancy['Dim1'] == 'Both sexes')
```

```python
filteredLifeExpectancy = lifeExpectancy[lifeExpectancy2019AndBothSexes]
```

```python
filteredLifeExpectancy['First Tooltip'].plot(kind='hist', bins = 10)
```

```python
filteredLifeExpectancy.sort_values(by=['First Tooltip'])
```

Eerste 5 rijen zijn dus de landen met de laagste levenverwachting. De laatste 5 rijen zijn de landen met de hoogste levensverwachting.
